```python
import matplotlib.pyplot as plt
plt.close()
# Data for the chart
previous_count = 18668
current_count = 13682831

# Plotting the data
fig, ax = plt.subplots()
ax.set_aspect('equal')

# Plotting circles to represent the data proportionally
previous_circle = plt.Circle((0, 0), radius=(previous_count ** 0.5), color='#87CEEB', alpha=0.7, label='Previous Count: 6,000')
current_circle = plt.Circle((0, 0), radius=(current_count ** 0.5), color='#1E90FF', alpha=0.7, label='Current Count: 600,000')

# Add circles to plot
ax.add_artist(previous_circle)
ax.add_artist(current_circle)

# Set limits for better visualization
ax.set_xlim(-current_count ** 0.5 * 1.2, current_count ** 0.5 * 1.2)
ax.set_ylim(-current_count ** 0.5 * 1.2, current_count ** 0.5 * 1.2)

# Annotations and legend
ax.annotate('ICTV VMR: 18668', xy=(previous_count ** 0.5, 0), xytext=(previous_count ** 0.5 * 2, 0),
            arrowprops=dict(facecolor='black', arrowstyle='->'))
ax.annotate('NCBI Virus: 13682831', xy=(current_count ** 0.5, 0), xytext=(current_count ** 0.5 * 1.2, -current_count ** 0.5 * 0.5),
            arrowprops=dict(facecolor='black', arrowstyle='->'))
# ax.legend(loc='lower right')

# Hide the axes
ax.axis('off')

# Show the plot
plt.show()
```
